### Analysing the frequencies of different weathers and plotting a line plot to show the same 

```python
Weather=data['Weather Conditions'].value_counts().reset_index()
Weather.columns=['Weather','Count']
Weather.head(25)
```

```python
sns.set_style('darkgrid')
plt.figure(figsize=(15,8))
plt.xticks(rotation=90)
sns.lineplot(x=Weather['Weather'],y=Weather['Count'],data=Weather,color='darkblue')
```

```python
plt.figure(figsize=(15,8))
plt.xticks(rotation=90)
plt.title('Weather vs Temperature', weight='bold', fontsize=12)
sns.barplot(x=data['Weather Conditions'],y=data['Temp_C'],ci=None)
```

```python
plt.figure(figsize=(15,8))
plt.xticks(rotation=90)
plt.title('Weather vs Pressure')
sns.lineplot(x=data['Weather Conditions'],y=data['Press_kPa'],data=data)
```

```python
plt.figure(figsize=(20,20))
sns.jointplot(x=data['Rel Hum_%'],y=data['Temp_C'],color='darkblue')
```
